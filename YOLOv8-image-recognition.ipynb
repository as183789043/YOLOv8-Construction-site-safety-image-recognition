### 匯入RoboFlow資料集

```python
rf = Roboflow(api_key=userdata.get('key'))
project = rf.workspace("roboflow-universe-projects").project("construction-site-safety")
dataset = project.version(30).download("yolov8")
```

### 將資料用來訓練模型

載入YOLO8模型的大小選擇參考連結 https://docs.ultralytics.com/tasks/detect/
image.png

```python
# Load a model
model = YOLO("yolov8m.pt")  # load a pretrained model (recommended for training)

# Use the model
model.train(data="/content/Construction-Site-Safety-30/data.yaml", epochs=100,patience=50)  # train the model
metrics = model.val()  # evaluate model performance on the validation set
```

### 使用資料集外照片供模型判斷

```python
model.predict('/content/Construction-Site-Safety-30/-1079-_png_jpg.rf.77459a818bed5a44927d46bdb976c661.jpg', save=True)
```
